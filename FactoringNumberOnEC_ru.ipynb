```python
from time import time
from random import randint


# сложение точек эллиптической кривой
# возвращает: 
# (1) полученная в результате сложения точка
# (2) знаменатель коэффициента наклона касательной при сложении точек
def points_addition(a1, b1, a2, b2, a, n):
    if a1 == a2 and b1 == b2:
        try:
            znam = (2 * b1) % n
            m = ((3 * a1**2 + a) / (znam)) % n
        except ZeroDivisionError:
            if (2 * b2) % n == 0:
                return 1, 0, 1
            else:
                znam = (2 * b1) % n
                return 'Error', 'Error', znam
    else:
        try:
            znam = (a2 - a1) % n
            m = ((b2 - b1) / (znam)) % n
        except ZeroDivisionError:
            if (a2 - a1) % n == 0:
                return 1, 0, 1
            znam = (a2 - a1) % n
            return 'Error', 'Error', znam
    res_x = (m**2 - a1 - a2) % n
    res_y = (m * (a1 - res_x) - b1) % n
    return res_x, res_y, znam


# умножение точки эллиптической кривой на число
# возвращает: 
# (1) полученная в результате умножения точка
# (2) список знаменателей коэффициентов наклона касательных,
# использованных при сложении
def point_number_multiplication(a, b, k, A, n):
    if k == 0:
        return 0, 0, [1]
    elif k == 1:
        return a, b, [1]
    elif k == 2:
        x, y, lambda_1 = points_addition(a, b, a, b, A, n)
        return x, y, [lambda_1]
    else:
        lambda_1_list = []
        x, y = a, b
        for i in range(1, k):
            x, y, lambda_1 = points_addition(x, y, a, b, A, n)
            lambda_1_list.append(lambda_1)
        return x, y, lambda_1_list


# разложение числа на множители, алгоритм на эллиптической кривой
# принимает на вход:
# (1) число n, которое нужно разложить
# (2) размер базы m
# возвращает:
# (1) нетривиальный делитель числа n
def factorization_on_ec(n, m):
    while True:
        # 1 генерация эллиптической кривой и случайной точки Q
        flag = 0
        a = randint(1, n // 2)
        b = randint(1, n // 2)
        for q_i_y in range(2, n - 1):
            for q_i_x in range(2, n - 1):
                if (q_i_y**2 - q_i_x**3 - a * q_i_x - b) % n == 0:
                    flag = 1
                    break
            if flag == 1:
                break
        print('Эллиптическая кривая:\na = {}\nb = {}\nn = {}'.format(a, b, n))
        print('Точка Q = ({}, {})'.format(q_i_x, q_i_y))
        iter_count = 0
        # 2
        i = 0
        # 3
        while i < m:
            p_i = Primes()[i]
            # 4
            i += 1
            alpha_i = int(0.5 * (log(n, 2).n())/(log(p_i, 2).n()))
            j = 0
            # 5
            while j < alpha_i:
                iter_count += 1
                q_i_x, q_i_y, lambda_list = point_number_multiplication(q_i_x, q_i_y, p_i, a, n)
                gcd_list = [gcd(n, lambda_) for lambda_ in lambda_list]
                d = [gcd_ for gcd_ in gcd_list if gcd_ > 1 and gcd_ < n]
                if d != []:
                    print('Разложение найдено, количество итераций = {}'.format(iter_count))
                    return d[0]
                j += 1


if __name__ == '__main__':
    # генерация случайного непростого числа для разложения
    while True:
        n = randint(2**10, 2**15)
        if not is_prime(n) and n % 2 != 0:
            break
    timer = time()
    d = factorization_on_ec(n, 10)
    print('Нетривиальный делитель d = {} найден за {} сек'.format(d, time() - timer))
    
```
